Dando continuidad con la primera entrega del proyecto, en esta oportunidad el estudiante debe realizar las siguientes validaciones utilizando la sentencia condicional IF.

Si el empleado es mayor de 55 años disfrutará de un bono de prepensión correspondiente al 5% de su sueldo básico.

Si el empleado es casado y tiene hijos se le otorgará un paseo cada diciembre

Si el sueldo básico está entre 1000000 y 1500000 tendrá una comisión del 2% sobre el valor del sueldo; Si el sueldo básico está entre 1500001 y 2000000 tendrá una comisión del 5% sobre el valor del sueldo; para todos los demás casos no habrá comisión.

Si el empleado trabajó más de 20 días al mes y su sueldo es menor a 1000000 tendrá derecho a un bono de alimentación.

```python
No_Identificacion = input("Digite su número de identificación: ")
Nombres = input("Digite su nombre: ")
Apellidos = input("Digite su apellido: ")
Direccion = input("Digite su direccion: ")
Telefono = input("Digite su número de telefono: ")
Edad = int(input("Digite su edad: "))
Estado_Civil = input("Digite su Estado Civil: ")
Numero_Hijos = int(input("Digite cuantos hijos tiene: "))
Fecha_Contr = input("Digite la fecha en la que ingreso a la compañia: ")
Sueldo_Basico = float(input("Didige su suedo básico: "))
Diaslaboradosmes = int(input("Digite los días que laboro en el mes: "))

if Edad > 55:
  bono = Sueldo_Basico * 0.05
  print(f"Tienes un bono adicional al sueldo por = ${bono:,.0f}")
else:
  print("No tienes bono adicional")
  
if Estado_Civil == "Casado" or "Casada" and Numero_Hijos > 0:
  print("Obtines un paseo en el mes de Diciembre junto con tu familia")
else:
  print("No tienes acceso al paseo en el mes de diciembre")
if Sueldo_Basico >= 1000000 or Sueldo_Basico <= 1500000:
  comision1 = Sueldo_Basico * 0.02
  print(f"Tienes una comisión adicional de = ${comision1:,.0f}")
elif Sueldo_Basico > 1500000 and Sueldo_Basico < 2000000:
  comision2 = Sueldo_Basico * 0.05
  print(f"Tienes una comisión adicional de = ${comision2:,.0f}")
else:
  print("No tienes comisión adicional a tu sueldo")
if Diaslaboradosmes > 20 and Sueldo_Basico < 1000000:
  print("Tienes derecho a bono de alimentación")
else:
  print("No tienes derecho al bono de alimentación")
```
