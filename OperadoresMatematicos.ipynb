```python
num1 = int(input("Digite el número 1: "))
num2 = int(input("Digite el número 2: "))

suma = num1 + num2
resta = num1 - num2
multi = num1 * num2
divi = num1 / num2

print("\nEl resultado de la SUMA es = {}". format(suma))
print("\nEl resultado de la RESTA es = {}". format(resta))
print(f"\nEl resultado de la MULTIPLICACIÓNN es = {multi}")
print(f"\nEl resultado de la DIVISIÓN es = {divi}")
```

Se recibe una venta por concepto de un producto se requiere aplicar el IVA (19%) y conoccer el total a pagar con el IVA incluido


```python
venta = float(input("Cuanto fue el total de la compra?: "))

iva = venta * 0.19
total = venta + iva

print("El Iva es = ${} y el total a pagar es= ${}".format(iva, total))
```

```python
venta = float(input("Cuanto fue el total de la compra?: $"))

iva = venta * 0.19
total = venta + iva

print("El Iva es = ${:,.0f} y el total a pagar es= ${:,.0f}".format(iva, total))
```
