Necesitamos saber si el usuario es alto o chaparro. Chaparro es aquel que mide cuabdi mucho 1.50 mts. Debe Preguntar si desea realizarlo de nuevo.

```python
def valeestatura(est):
  if est > 1.50:
    print("Eres una persona ALTA")
  else:
    print("Eres una persona CHAPARRA")

resp = "si"
while resp == "si" or resp =="s":
  estatura = float(input("Digite su estatura: "))
  valeestatura(estatura)
  resp = input("Desea realizar otra comparación? ")
```
