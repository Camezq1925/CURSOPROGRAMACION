Un vendedor recibe un sueldo base más un 10% extra por
comisión de sus ventas, el vendedor desea saber cuánto
dinero obtendrá por concepto de comisiones por las tres
ventas que realiza en el mes y el total que recibirá en
el mes tomando en cuenta su sueldo base y comisiones.

```python
#35200000 
venta1 = float(input("Se vendió carro por: $"))

comisión1 = venta1 * 0.10

print("\nEl valor de la comisión es = ${:,.0f}".format(comisión1))

#59750000
venta2 = float(input("Se vendió camioneta por: $"))

comisión2 = venta2 * 0.10

print("\nEl valor de la comisión es = ${:,.0f}".format(comisión2))

#17270000
venta3 = float(input("Se vendió carro por: $"))

comisión3 = venta3 * 0.10

print("\nEl valor de la comisión es = ${:,.0f}".format(comisión3))

#2500000
sueldo = float(input("El vendedor tiene un sueldo base de: $"))

totalcomisiones = comisión1 + comisión2 + comisión3
totalsuedo = sueldo + totalcomisiones

print("\nEl total de las comisiones es = ${:,.0f} y el total del sueldo que recibió el vendedor es = ${:,.0f}".format(totalcomisiones, totalsuedo))
```

Una tienda ofrece un descuento del 15% sobre el total
de la compra y un cliente desea saber cuánto deberá
pagar finalmente por su compra.

```python
#68350
compra = float(input("Cliente compra una maleta por: $"))

descuento = compra * 0.15
totalcompra = compra - descuento

print("\nEl descuento sobre la malesta es de = ${:,.0f} y el total a pagar es= ${:,.0f}".format(descuento, totalcompra))
```

Un alumno desea saber cuál será su calificación final en
la materia de Algoritmos. Dicha calificación se compone
de tres exámenes parciales.

```python
#4.3; 3.2; 4.8
parcial1 = float(input("Camilo sacó en su primer parcial: "))
parcial2 = float(input("Camilo sacó en su segundo parcial: "))
parcial3 = float(input("Camilo sacó en su tercer parcial: "))

totalnota = (parcial1 + parcial2 + parcial3) / 3

print(f"\nLa nota final que tuvo Camilo es = {totalnota}")
```

Un maestro desea saber qué porcentaje de hombres y que
porcentaje de mujeres hay en un grupo de estudiantes.

```python
#4325 Hombres y 7681 Mujeres
hombres = float(input("En una región se realizó un censo el cual nos indica que hay un total de hombres de: "))
mujeres = float(input("En una región se realizó un censo el cual nos indica que hay un total de mujeres de: "))

totalpoblacion = hombres + mujeres

porcentajehombres = (hombres / totalpoblacion) * 100
porcentajemujeres = (mujeres / totalpoblacion) * 100

print(f"\nEl porcentaje de hombres en la región es = {porcentajehombres}% y el procentaje de mujeres en la región es ={porcentajemujeres}%")
```
