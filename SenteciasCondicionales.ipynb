Se necesita un sistema para un supermercado, el cual dará un 10% de descuento a las personas que compren más de $1000000 de lo contrario no dará ningún descuento, al cliente se le debe dar el total a pagar. Condicional.

```python
compra = float(input("Digite el valor de la compra: $"))

if compra > 1000000:
  descu = compra * 0.10
  total = compra - descu
  print("El total del descuento = {:,.0f} y el total a pagar es 0 ${:,.0f}".format(descu, total))
else:
  print("No tiene descuento, el total a pagar es = ${:,.0f}".format(compra))
  print(f"No tiene descuento, el total a pagar es = ${compra:,.0f}")
```

Se necesita un sistema que reciba tres calificaciones parciales de un alumno y en base a estas darle su promedio donde si el promedio es menor a 3.0 se le dirá que esta reprobado, en caso contrario el mensaje será aprobado. Condicional.

```python
cal1 = float(input("Digite la calificación 1: "))
cal2 = float(input("Digite la calificación 2: "))
cal3 = float(input("Digite la calificación 3: "))

prom = round((cal1 + cal2 + cal3)/3, 2)

if prom < 3.0:
  estado = "Reprobado--"
else:
  estado = "Aprobado++"
print(f"El promedio es = {prom} y el estado es = {estado}")
```
