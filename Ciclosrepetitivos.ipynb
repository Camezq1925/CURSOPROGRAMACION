Escribir un programa que realice la tabla de multiplicar de un numero digitado por el teclado

```python
num = int(input("Digite el numero de la tabla de multipicar: "))

for i in range(11):
  print(i, " x ", num, " = ", i * num)
```

Escriba un programa que sume los numero pares del 0 al 10 y saque el total de la suma

```python
#variables acumuladoras
acu = 0

for i in range(0,11,2):
  #print(i)
  acu = acu + i
print(f"El total de la suma de los numero pares del 0-10 es = {acu}")
```

Ciclo while

```python
x = 30
while x <= 10:
  print(x)
  x += 1
```

```python
x = 7
while x >= 5:
  print(x)
  x = x - 1
```

```python
x = 0
while x <= 15:
  x += 3
  print(x)
```

```python
a = 5
b = 100
c = 10

while a * c < b:
  a = a + 5
  c = c - 1
print("a={}, b={}, c={}".format(a,b,c))
```

```python
resp = "si"
while resp == "si":
  a = 5
  b = 100
  c = 10

  while a * c < b:
    a = a + 5
    c = c - 1
  print("a={}, b={}, c={}".format(a,b,c))
  resp = input("desea realizarlo de nuevo?(si/no)")
```

Se requiere un software que calcule promedio de tres materias si es >=3.0 el estado es apruba sino reprueba

```python
resp = "si"
while resp == "si" or resp == "SI" or resp == "yes":
  cal1 = float(input("Digite la calificación 1: "))
  cal2 = float(input("Digite la calificación 2: "))
  cal3 = float(input("Digite la calificación 3: "))

  prom = round((cal1 + cal2 + cal3)/3, 2)

  if prom >= 3.0:
   estado = "Aprueba"
  else:
    estado = "Reprueba"
  print(f"El promedio es = {prom} y el estado es = {estado}")
  resp = input("Desea ingresar otras 3 calificaciones?")
```
